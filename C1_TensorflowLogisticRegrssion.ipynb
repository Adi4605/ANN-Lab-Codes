```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
print("Neural Network(XOR)")
x1=np.array([[0,0],[0,1],[1,0],[1,1]])
y1=np.array([0,1,1,0])

model1=Sequential([
    Dense(4,input_dim=2,activation='relu'),
    Dense(1,activation='sigmoid')
])
model1.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model1.fit(x1,y1,epochs=5000,verbose=0)

print("Predictions: ")
print(model1.predict(x1))
```

```python
print("\nLogistic Regression")

X2 = np.array([[0],[1],[2],[3]])
y2 = np.array([0,0,1,1])

model2 = Sequential([
    Dense(1, input_dim=1, activation='sigmoid')
])

model2.compile(optimizer='adam',
               loss='binary_crossentropy',
               metrics=['accuracy'])

model2.fit(X2, y2, epochs=500, verbose=0)
```

```python
loss, acc = model2.evaluate(X2, y2, verbose=0)
print(model2.predict(X2))
print("Accuracy:", acc)
```
