```python
import numpy as np
import matplotlib.pyplot as plt

# Activation functions
def linear(x):
    return x

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.1):
    return np.where(x > 0, x, alpha * x)

# Input range
x = np.linspace(-10, 10, 500)

# Create subplots
plt.figure(figsize=(8, 6))

plt.subplot(2, 3, 1)
plt.plot(x, linear(x), color="black")
plt.title("Linear")
plt.grid(True)

plt.subplot(2, 3, 2)
plt.plot(x, sigmoid(x), color="blue")
plt.title("Sigmoid")
plt.grid(True)

plt.subplot(2, 3, 3)
plt.plot(x, tanh(x), color="green")
plt.title("Tanh")
plt.grid(True)

plt.subplot(2, 3, 4)
plt.plot(x, relu(x), color="red")
plt.title("ReLU")
plt.grid(True)

plt.subplot(2, 3, 5)
plt.plot(x, leaky_relu(x), color="purple")
plt.title("Leaky ReLU")
plt.grid(True)

plt.tight_layout()
plt.show()
```
