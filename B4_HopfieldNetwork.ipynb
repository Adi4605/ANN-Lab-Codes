```python
import numpy as np
```

```python
p=int(input("Enter number of vectors: "))
n=int(input("Enter the size of each vetcor: "))

patterns=[]

for i in range(p):
    print(f"\nEnter vectors {i+1} elements (1 or -1): ")
    vec = np.array(list(map(int,input().split())))
    patterns.append(vec)
```

```python
w=np.zeros((n,n)) 
# Training Hopfield Network
for vec in patterns:
    w+= np.outer(vec,vec)
np.fill_diagonal(w,0)   #Remove self-connections
print("\nWeight Matrix:\n")
print(w)
```

```python
# Testing
print("\nEnter test vector: ")
test=np.array(list(map(int,input().split())))
print("\nTest Vector: ")
print(test)

#Recall process
result=np.dot(w,test)
result=np.where(result>=0,1,-1)
print("\nRecovered Vector:")
print(result)
```
