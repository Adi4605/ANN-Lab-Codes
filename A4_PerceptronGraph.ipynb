```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=int(input("Enter the number of training points: "))
X=[]
Y=[]
for i in range(n):
    x1=float(input(f"Enter x1 for point {i+1}: "))
    x2=float(input(f"Enter x2 for point {i+1}: "))
    label = int(input("Enter the class(0 or 1): "))

    X.append([x1,x2])
    Y.append(label)

X=np.array(X)
Y=np.array(Y)
```

```python
w=np.zeros(2)
b=0
lr=0.1

#Training
for epoch in range(100):
    for i in range(len(X)):
        net=np.dot(X[i],w)+b

        #Activation Function
        if net>=0:
            output=1
        else:
            output=0

        error=Y[i]-output

        w+=lr*error*X[i]
        b+=lr*error

print("Training Completed")
print("Weights: ",w)
print("Bias: ",b)
```

```python
#Graph
for i in range(len(X)):
    if Y[i]==0:
        plt.scatter(X[i][0,X[i][1],marker='o',label='Class 0')
    else:
        plt.scatter(X[i][0,X[i][1],marker='x',label='Class 1')

# Decision Boundary
x_vals = np.linspace(min(X[:,0])-1, max(X[:,0])+1, 100)

if w[i]!=0:
    y_vals = -(w[0]x_vals+b)/w[1]
    plt.plot(x_vals, y_vals)

plt.xlabel("X1")
```
