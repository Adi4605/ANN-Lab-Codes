```python
import numpy as np
```

```python
class Perceptron:
    def __init__ (self, lr=0.01, epochs=20):
        self.weight = 0.0
        self.bias =0.0
        self.lr = lr
        self.epochs = epochs

    def activation(self,x):
        return 1 if x>0 else 0

    def predict(self,x):
        total = self.weight*x + self.bias
        return self.activation(total)

    def train(self,x,y):
        for _ in range(self.epochs):
            for xi, target in zip(x,y):
                prediction = self.predict(xi)
                error = target - prediction
                self.weight += self.lr*xi*error
                self.bias += self.lr*error
```

```python
digits = list(range(10))

x = np.array([d%2 for d in digits])
y = np.array([d%2 for d in digits])
```

```python
# Model
p=Perceptron(lr=0.01, epochs=20)
p.train(x,y)
```
