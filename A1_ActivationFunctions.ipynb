```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,100)
```

```python
linear = x
sigmoid = 1/(1+np.exp(-x))
tanh = np.tanh(x)
relu = np.maximum(0,x)
leaky_relu = np.where(x>0, x, 0.01*x)
```

```python
plt.figure()
plt.plot(x,linear)
plt.title("Linear Function")
plt.grid()
plt.show()

plt.figure()
plt.plot(x,sigmoid)
plt.title("Sigmoid Function")
plt.grid()
plt.show()

plt.figure()
plt.plot(x,tanh)
plt.title("Tanh Function")
plt.grid()
plt.show()

plt.figure()
plt.plot(x,relu)
plt.title("ReLU Function")
plt.grid()
plt.show()

plt.figure()
plt.plot(x,leaky_relu)
plt.title("Leaky ReLU Function")
plt.grid()
plt.show()
```
