```python
# McCulloch-Pitts neuron model for ANDNOT function
def mcculloch_pitts_andnot(A,B):
    #weights
    w1 = 1      #weight for A
    w2 = -1     #weight for B (inhibitory)
    
    #threshold
    theta = 1
    
    # neuron activation
    net_input = (w1 * A) + (w2 * B)
    
    if net_input >= theta:
        return 1
    else:
        return 0
    
# Test all input combinations
print("A B | Output (A AND NOT B)")
print("--------------------------")
    
inputs = [(0,0), (0,1), (1,0), (1,1)]
    
for A, B in inputs:
    output = mcculloch_pitts_andnot(A,B)
    print(F"{A} {B} | {output}")
    
```
