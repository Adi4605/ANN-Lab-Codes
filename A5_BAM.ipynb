```python
import numpy as np
```

```python
x1 = np.array([1,-1,1])
x2 = np.array([-1,1,-1])

y1 = np.array([1,-1])
y2 = np.array([-1,1])

W = np.outer(x1,y1) + np.outer(x2,y2)
print("Weight Matrix:\n", W)
```

```python
def sign(x):
    return np.where(x>=0, 1,-1)
```

```python
# Test BAM
# Forward: x → y
print("\nForward Recall:")
print("x1 ->", sign(np.dot(x1, W)))
print("x2 ->", sign(np.dot(x2, W)))

# Backward: y → x
print("\nBackward Recall:")
print("y1 ->", sign(np.dot(W, y1)))
print("y2 ->", sign(np.dot(W, y2)))
```
